# POLYNOMIAL LINEARREGRESSION

# IMPORTING THE LIBRARIES

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# IMPORTING THE DATASET

```python
dataset = pd.read_csv('Position_Salaries.csv')
x = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

# TRAINING THE LINEAR REGRESSION WITH THE WHOLE DATASET

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x, y)
```

# TRAINING THE POLYNOMIAL REGRESSION WITH THE WHOLE DATASET

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=7)
x_poly = poly_reg.fit_transform(x)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly, y)
```

# VISUALING THE LINEAR REGRESSION


```python
plt.scatter(x, y, color = 'red')
plt.plot(x, lin_reg.predict(x), color = 'blue')
plt.xlabel("Level")
plt.ylabel("Salary")
plt.title("Salary against level at  (LinearRegression)")
plt.show()
```

# VISIONALIZING THE POLYNOMIAL REGRESSION

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, lin_reg_2.predict(x_poly), color = 'blue')
plt.title("Salary against level at (PolynomialRegression)")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

# Visualizing the polynomial in smoother(For higher resolution and soother curve)

```python
# x = np.array(x, dtype=float)
x_grid = np.arange(min(x), max(x), 0.1)
x_grid = x_grid.reshape(len(x_grid), 1)
plt.scatter(x, y, color = 'red')
plt.plot(x_grid, lin_reg_2.predict(poly_reg.fit_transform(x_grid)), color = 'blue')
plt.title("Salary against level at (PolynomialRegression)")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

# PREDICTING USING THE LINEAR REGRESSION

```python
lin_reg.predict([[6.5]])
```

Taking the linear regression model gives the very very wrong answer

## PREDICTING THE RESULT USING POLYNOMIAL REGRESSION

```python
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
